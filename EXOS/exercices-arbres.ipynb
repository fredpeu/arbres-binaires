# Exercice 4:   ` Retour sur la RPN`

Dans notre chapitre sur les piles, nous avions évoqué la notation polonaise inversée (RPN) qui permet d´évaluer une expression arithmétique sans ambiguïté vis-à-vis des priorités. 

La fonction `rpn` qui permet d´évaluer une expression arithmétique en notation polonaise inversée est donnée ci-dessous. Elle prend en paramètre une liste contenant l´expression arithmétique. L´expression en RPN est dite **postfixée**. Nous allons comprendre pourquoi.

```python
# on importe le fichier qui contient la classe Pile
from pile import *

def operateur(c):
    if c=='+' or c=='-'or c=='*'or c=='/':
        return True
    else:
        return False
    
def calculer(a,b,op):
    if op=='+':return a+b
    if op=='-':return a-b
    if op=='*':return a*b
    if op=='/':
        if b!=0:return a/b
        else: print("erreur de division par zero")
        
def rpn(expression):
    p=Pile() # creer_pile
    for e in expression:
        if operateur(e):# si un opérateur a été trouvé
            a=p.sommet.element  # on stocke l'élément de sommet de P dans la variable a
            p.depiler()  # on dépile P
            b=p.sommet.element # on stocke l'élément de sommet de P dans la variable b
            p.depiler()
            r=calculer(b,a,e) # on calcule l'opération trouvée
            p.empiler(r) # on empile P avec le résultat              
        else: 
            p.empiler(int(e)) # on empile le nombre de l'expression             
    return p.sommet.element
```
